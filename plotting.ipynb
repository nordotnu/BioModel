## Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

data = pd.read_csv("data/plt/30hz-in-WLc-RMSc.csv")
wl = data["f7"]
rms = data["f3"]

data = pd.read_csv("data/plt/raw-in-rms-c .csv")
raw = data["f3"]
raw.size
t_raw = np.arange(0, 3, (3 / raw.size))
t_f = np.arange(0, 3, (3 / rms.size))

```

```python
fig, axs = plt.subplots()
axs.set_title("Raw sEMG Signal")
axs.plot(t_raw, raw, linewidth=0.5, markersize=12)
axs.set_xlabel("Time")
axs.set_ylabel("Amplitude")

fig.set_figwidth((10))

plt.show()


fig, axs = plt.subplots()
axs.set_title("Extracted Features from Signal (N=20)")
axs.plot(t_f, wl, linewidth=1, markersize=12, label='Wave Length (WL)')
axs.legend()
axs.plot(t_f, rms, linewidth=1, markersize=12, label='Root Mean Squared (RMS)')
axs.legend()
axs.set_xlabel("Time")
axs.set_ylabel("Amplitude")

fig.set_figwidth((10))
plt.show()
```
